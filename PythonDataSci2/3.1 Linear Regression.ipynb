# Chapter 3 - Regression Models
## Segment 1 - Simple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from pylab import rcParams

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,8
```

```python
rooms = 2*np.random.rand(100, 1) + 3
rooms[1:10]
```

```python
price = 265 + 6*rooms +abs(np.random.rand(100, 1))
price[1:10]
```

```python
plt.plot(rooms, price, 'r^')
plt.xlabel("# of Rooms, 2019 Average")
plt.ylabel('2019 Average Home, 1000s USD')
plt.show()
```

```python
X = rooms
y = price

LinReg = LinearRegression()
LinReg.fit(X,y)
print(LinReg.intercept_, LinReg.coef_)
```

*Simple Algebra*
- y = mx + b
- b = intercept = 265.6

*Estimated Coefficients*
- LinReg.coef_ = [5.97] Estimated coefficients for the terms in the linear regression problem.

```python
### how well our model performed (r^2)

print(LinReg.score(X,y))
```
